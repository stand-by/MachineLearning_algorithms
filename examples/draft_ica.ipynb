```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy import linalg
#https://en.wikipedia.org/wiki/FastICA
#http://math.stackexchange.com/questions/695194/is-my-implementation-of-fastica-right
```

```python
sig1 = np.sin(np.linspace(0,20,200))
sig2 = np.sinc(np.linspace(0,20,200))
plt.plot(sig1)
plt.plot(sig2)
```

```python
signals = np.array([sig1,sig2])
A = np.array([[10,50],[0.3,2]])
mixed_signals = np.dot(A,signals)

plt.plot(mixed_signals[0,:])
plt.plot(mixed_signals[1,:])
plt.show()

recovered = np.dot(np.linalg.pinv(A),mixed_signals)
plt.plot(recovered[0,:])
plt.plot(recovered[1,:])
plt.show()
```

```python
X = mixed_signals
X = X - np.mean(X, axis=1).reshape((2,1))
covariance = np.dot(X,X.T)
eig_stuff = linalg.eig(covariance)
cov_eigvals,cov_eigvecs = eig_stuff[0].astype('d'),eig_stuff[1]
E = cov_eigvecs
D = np.diag(cov_eigvals)
X = E.dot(linalg.sqrtm(linalg.pinv(D))).dot(E.T).dot(X)
```

```python
def g(u):
    return u*np.exp(-u**2/2.0)
def dg(u):
    return (1-u**2)*np.exp(-u**2/2.0)

m = X.shape[1]
W = []

#for p in range(2):
w = np.random.randn(2)
w_prev = np.zeros(2)
for _ in range(500000):
    w=w.reshape(2,1)
    w_prev = w
    
    w = X.dot(g(np.dot(w.T,X)).T)/float(m) - (dg(np.dot(w.T,X)).dot(np.ones((m,1)))[0][0]*(w))/float(m)
    w = w/linalg.norm(w)
W.append(w.flatten())

v = np.random.randn(2)
v_prev = np.zeros(2)
for _ in range(500000):
    v=v.reshape(2,1)
    v_prev = v
    
    v = X.dot(g(np.dot(v.T,X)).T)/float(m) - (dg(np.dot(v.T,X)).dot(np.ones((m,1)))[0][0]*(v))/float(m)
    v = v - v.T.dot(w)*(w)
    v = v/linalg.norm(v)
W.append(v.flatten())
    
W = np.array(W)
```

```python
rec_sig1 = np.dot(w.T,X)[0]
rec_sig2 = np.dot(v.T,X)[0]

plt.plot(rec_sig1)
plt.plot(rec_sig2)
```
