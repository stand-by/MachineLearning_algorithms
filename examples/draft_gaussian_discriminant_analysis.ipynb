```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
def splitDataset(dataset, splitRatio):
	trainSize = int(len(dataset) * splitRatio)
	trainSet = []
	copy = list(dataset)
	while len(trainSet) < trainSize:
		index = np.random.randint(0,len(copy))
		trainSet.append(copy.pop(index))
	return (np.array(trainSet), np.array(copy))

X = np.loadtxt("datasets/students_admission.csv",delimiter=',')
#X,X_test = splitDataset(dataset,0.7)
X,y = X[:,:-1],X[:,-1]
#X_test,y_test = X_test[:,:-1],X_test[:,-1]

pos = X[y==1]
neg = X[y==0]
plt.plot(pos[:,0], pos[:,1], 'go')
plt.plot(neg[:,0], neg[:,1], 'ro')
```

```python
mean_pos = np.mean(pos,axis=0)
mean_neg = np.mean(neg,axis=0)
covariation = np.dot((X-np.mean(X,axis=0)).T, (X-np.mean(X,axis=0)))/float(X.shape[0])
covariation_pos = np.dot((X-mean_pos).T, (X-mean_pos))/float(pos.shape[0])
covariation_neg = np.dot((X-mean_neg).T, (X-mean_neg))/float(neg.shape[0])
plt.plot(pos[:,0], pos[:,1], 'go')
plt.plot(neg[:,0], neg[:,1], 'ro')
plt.plot(mean_pos[0], mean_pos[1], 'x', markersize=10)
plt.plot(mean_neg[0], mean_neg[1], 'x', markersize=10)
```

```python
def gaussian_density(x,mean,sigma):
    n = len(x)
    return np.exp(((x-mean).T.dot(np.linalg.pinv(sigma)).dot(x-mean))/-2.0)/((np.linalg.det(sigma)**(1/2.0))*(2*np.pi)**(n/2.0))

xlist = np.linspace(20.0, 110.0, 500)
ylist = np.linspace(20.0, 110.0, 500)
Xgrid, Ygrid = np.meshgrid(xlist, ylist)

Z = np.zeros((len(xlist),len(ylist)))
for i in range(len(xlist)):
    for j in range(len(ylist)):
        Z[i,j] = gaussian_density(np.array([xlist[i],ylist[j]]),mean_pos,covariation)
plt.contour(Xgrid, Ygrid, Z)

Z = np.zeros((len(xlist),len(ylist)))
for i in range(len(xlist)):
    for j in range(len(ylist)):
        Z[i,j] = gaussian_density(np.array([xlist[i],ylist[j]]),mean_neg,covariation)
plt.contour(Xgrid, Ygrid, Z)

plt.plot(pos[:,0], pos[:,1], 'go')
plt.plot(neg[:,0], neg[:,1], 'ro')
plt.plot(mean_pos[0], mean_pos[1], 'x', markersize=10)
plt.plot(mean_neg[0], mean_neg[1], 'x', markersize=10)
```

```python
def probabilities(input_vector):
    return np.array([gaussian_density(input_vector,mean_neg,covariation),gaussian_density(input_vector,mean_pos,covariation)])

def predict(input_vector):
    probs = probabilities(input_vector)
    if probs[0] > probs[1]: return 0
    else: return 1
    
predictions = []
for i in range(len(X)):
    predictions.append(predict(X[i]))
print "Accuracy on training set", np.mean(predictions==y)
```
